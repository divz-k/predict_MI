```python
#check fast api is running
import requests

url = "http://127.0.0.1:8000/docs"
response = requests.get(url)
print(response.status_code)  # Should be 200
```

```python
# make a prediction from excel file using fast api: local
import requests

url = "http://127.0.0.1:8000/predict_excel"
files = {'file': open('input_sample.xlsx', 'rb')}

response = requests.post(url, files=files)
print(response.text)         # see what the server actually returned
```

```python
# make a prediction from excel file using fast api: deployed on render.com

url = "https://predict-mi-app-latest.onrender.com/predict_excel"

# Replace with your Excel file path
files = {'file': open('input_sample.xlsx', 'rb')}

response = requests.post(url, files=files)

print(response.json())
```
